1.	Read the CSV file containing the sales data. - done
2.	Remove any rows with missing or invalid data. - done

```python
import pandas as pd
from datetime import datetime as dt
```

```python
sales_df = pd.read_csv('sales.csv')
```

```python
sales_df.head()
```

```python
sales_df.dtypes
```

```python
sales_df.describe(), sales_df.info()
```

```python
sales_df.isna().sum()
```

```python
cleaned_sales_df = sales_df.dropna()
```

```python
cleaned_sales_df.describe(), cleaned_sales_df.info()
```

3.	Calculate the total sales for each product category. - 

```python
cleaned_sales_df['sales'] = cleaned_sales_df['Order Quantity'] * cleaned_sales_df['Unit Price']
```

```python
cleaned_sales_df.head(10)
```

```python
total_sales_by_category = cleaned_sales_df.groupby('Product Category')['sales'].sum()
print(total_sales_by_category)
```

4.	Calculate the total sales for each customer.

```python
total_sales_for_customer = cleaned_sales_df.groupby('Customer Name')['sales'].sum()
```

```python
total_sales_for_customer
```

5.	Calculate the total sales for each day.

```python
total_sales_per_day = cleaned_sales_df.groupby('Order Date')['sales'].sum().reset_index()
```

```python
total_sales_per_day
```

6.	Calculate the average order value for each customer.

```python

avg_sales_per_customer = cleaned_sales_df.groupby('Customer Name')['sales'].mean().reset_index()
```

```python
avg_sales_per_customer
```

7.	Create a new column called "Order Month" that contains the month (in string format) of each order date.

```python
cleaned_sales_df['Order Date'] = pd.to_datetime(cleaned_sales_df['Order Date'])
cleaned_sales_df.dtypes
```

```python
cleaned_sales_df['Order Month'] = cleaned_sales_df['Order Date'].dt.strftime('%B')
```

```python
cleaned_sales_df.head()
```

8.	Save the cleaned and transformed data to a new CSV file.

```python
# dropping sales column
cleaned_sales_df.drop(columns=['sales'])
```

```python
cleaned_sales_df.to_csv('cleaned sales df.csv', index=False)
```
